```python
from bokeh.plotting import figure, show, row, column
from bokeh.io import output_notebook
from bokeh.util.hex import hexbin
from bokeh.transform import linear_cmap
from bokeh.palettes import all_palettes
from bokeh.layouts import layout
from bokeh.models import Slider
import numpy as np

output_notebook()
```

```python
x = np.linspace(-20 * np.pi, 20 * np.pi, 1000)
y_sin = np.sin(x)
y_cos = np.cos(x)

data = {'x': x, 'y_sin': y_sin, 'y_cos': y_cos}
```

```python
fig = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Mój mega wykres', sizing_mode = 'stretch_width', height = 300)
fig.toolbar.logo = None
fig.toolbar.autohide = True
fig.grid.grid_line_dash = (6, 5)

fig.line(x, y_sin)
fig.circle(x, y_sin, color = 'red', radius = 0.1 * np.log(np.abs(x)), fill_alpha = 0.5)
show(fig)
```

```python
fig1 = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Mój mega wykres', sizing_mode = 'stretch_width', height = 300)
fig1.toolbar.logo = None
fig1.toolbar.autohide = True
fig1.grid.grid_line_dash = (6, 5)

fig1.line(x, y_sin)

fig2 = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Mój mega wykres', sizing_mode = 'stretch_width', height = 300)
fig2.toolbar.logo = None
fig2.toolbar.autohide = True
fig2.grid.grid_line_dash = (6, 5)

fig2.circle(x, y_sin, color = 'red', radius = 0.1 * np.log(np.abs(x)), fill_alpha = 0.5)

show(column(fig1, fig2, sizing_mode = 'stretch_width'))
```

```python
fig = figure(x_range = (-np.pi, np.pi), x_axis_label = 'x', y_axis_label = 'y', title = 'Mój mega wykres', sizing_mode = 'stretch_width', height = 300)
fig.toolbar.logo = None
fig.toolbar.autohide = True
fig.grid.grid_line_dash = (6, 5)

fig.line('x', 'y_sin', source = data)
fig.circle(x, y_sin, color = 'red', radius = 0.1 * np.log(np.abs(x)), fill_alpha = 0.5)
show(fig)
```

```python
n_points = 100000
scale = 0.5
xs = -1 + 2 * np.random.random(n_points)
ys = xs + np.random.normal(scale = scale, size = n_points)
data = {'xs': xs, 'ys': ys}

points_slider = Slider(start = 1000, end = 1000000, step = 1000, value = n_points, title = 'Number of points', width = 300)
scale_slider = Slider(start = 0.01, end = 1, step = 0.01, value = scale, title = 'Scale', width = 300)

f1 = figure(match_aspect = True)
f1.circle('xs', 'ys', source = data, alpha = 0.1)

binned_data = hexbin(data['xs'], data['ys'], 0.01)
cmap = linear_cmap('counts', 'Turbo256', 0, max(binned_data['counts']))
f2 = figure(match_aspect = True)
f2.background_fill_color = all_palettes['Turbo'][256][0]
f2.grid.visible = False
f2.hex_tile(size = 0.01, source = binned_data, fill_color = cmap, line_color = None)

#show(column(points_slider, scale_slider))
#show(row(f1, f2, sizing_mode = 'stretch_width'))
show(layout([column(points_slider, scale_slider), row(f1, f2)], sizing_mode = 'stretch_width'))
```
